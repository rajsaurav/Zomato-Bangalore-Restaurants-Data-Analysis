# Cleaning Rate Column

```python
df['rate'].unique()
```

```python
import string
```

```python
def cleanrate(value):
    if (value=='NEW' or value == '-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(cleanrate)
df['rate'].head()
```

```python
df
```

```python
df.rate.isnull().sum()
```

# Filling Null Values in Rate Column 

```python
df['rate'].fillna(df['rate'].mean(), inplace= True)
df['rate'].isnull().sum()
```

# Fixing Values in Approx cost column

```python
def removecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(removecomma)
df['approx_cost(for two people)'].unique()
```

--> Analysing Restaurant Type Column

```python
df['rest_type'].value_counts()
```

```python
type_restaurant=df['rest_type'].value_counts(ascending=False)
type_restaurant
```

```python
type_restaurantlessthan1000 = type_restaurant[type_restaurant<1000]
type_restaurantlessthan1000
```

```python
 def handle_res_type(value):
        if (value in type_restaurantlessthan1000):
            return 'Others'
        else:
            return value

df['rest_type'] = df['rest_type'].apply(handle_res_type)
df['rest_type'].value_counts()
```
